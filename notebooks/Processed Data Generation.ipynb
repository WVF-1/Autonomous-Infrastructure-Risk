```python
# Import necessary libraries

import pandas as pd
from pathlib import Path

from src.preprocessing import preprocess_dataset
from src.constants import DATA_PATH_RAW, DATA_PATH_PROCESSED

# Create the new dataset

def main():
    print("Loading raw dataset...")
    df = pd.read_csv(DATA_PATH_RAW)

    print("Preprocessing dataset...")
    df_processed, _ = preprocess_dataset(df)

    print("Ensuring data directory exists...")
    Path(DATA_PATH_PROCESSED).parent.mkdir(parents=True, exist_ok=True)

    print("Saving processed dataset...")
    df_processed.to_csv(DATA_PATH_PROCESSED, index=False)

    print("✅ Processed dataset created at:")
    print(DATA_PATH_PROCESSED)

if __name__ == "__main__":
    main()
```
